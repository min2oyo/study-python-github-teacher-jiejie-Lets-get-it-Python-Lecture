## 그래프 꾸미기


### 제목 삽입


```python
month = ["mar", "apr", "may", "jun", "jul"]
sales = [1, 5, 7, 3, 7]

plt.title("월별 판매 실적")
plt.plot(month, sales)  # x축, y축
plt.show()
```

### 그래프 색 변경


```python
month = ["mar", "apr", "may", "jun", "jul"]
sales = [1, 5, 7, 3, 7]

plt.title("월별 판매 실적")
plt.plot(
    month, sales, color="y"
)  # 색상코드: https://matplotlib.org/2.0.2/examples/color/named_colors.html
plt.show()
```

### 크기 변경


```python
month = ["mar", "apr", "may", "jun", "jul"]
sales = [1, 5, 7, 3, 7]

plt.figure(dpi=150)  # dpi: 1제곱인치 안의 픽셀 개수
# plt.figure(figsize=(8, 4))  # 인치단위	# (가로, 세로)

plt.title("월별 판매 실적")
plt.plot(month, sales, color="b")
plt.show()
```

### 범례

```python
month = ["mar", "apr", "may", "jun", "jul"]
sales = [1, 5, 7, 3, 7]

plt.title("월별 판매 실적")
plt.plot(month, sales, color="b", label="단위: 개수")  # 범례 내용 설정
plt.legend()  # 범례 표시
plt.show()
```

```python
month = ["mar", "apr", "may", "jun", "jul"]
sales = [1, 5, 7, 3, 7]

plt.title("월별 판매 실적")
plt.plot(month, sales, color="b", label="단위: 개수")  # 범례 내용 설정
plt.legend(loc="lower right")  # 범례 표시
plt.show()
```

### 축 이름

```python
month = ["mar", "apr", "may", "jun", "jul"]
sales = [1, 5, 7, 3, 7]

plt.title("월별 판매 실적")
plt.plot(month, sales, color="b", label="단위: 개수")  # 범례 내용 설정
plt.legend(loc="lower right")  # 범례 표시
plt.xlabel("월")
plt.ylabel("매출")
plt.show()
```
