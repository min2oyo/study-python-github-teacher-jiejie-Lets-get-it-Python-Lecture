## 막대그래프 그리기


```python
import matplotlib.pyplot as plt

sales = [1, 5, 7, 3, 7]
month = [3, 4, 5, 6, 7]

plt.bar(month, sales)
plt.show()
```

### 제목 넣고 그래프 색 바꾸기


```python
sales = [1, 5, 7, 3, 7]
month = [3, 4, 5, 6, 7]

plt.rc("font", family="Malgun Gothic")
plt.title("월별 판매 실적")
plt.bar(month, sales, color="b")
plt.show()
```

### x축에 range()로 연속 숫자 넣기


```python
sales = [1, 5, 7, 3, 7]

plt.rc("font", family="Malgun gothic")
plt.title("월별 판매 실적")
plt.bar(range(3, 8), sales, color="b")
plt.show()
```

```python
sales = [1, 5, 7, 3, 7]

plt.rc("font", family="Malgun gothic")
plt.title("월별 판매 실적")
plt.bar(range(2, 11, 2), sales, color="b")
plt.show()
```

### 막대 방향 전환하기

```python
sales = [1, 5, 7, 3, 7]

plt.rc("font", family="Malgun gothic")
plt.title("월별 판매 실적")
plt.barh(range(3,8), sales, color="b")
plt.show()
```

### 컬러맵 적용하기

```python
import random

height = []
weight = []

for i in range(100):
    height.append(random.randint(100, 200))
    weight.append(random.randint(20, 100))

plt.rc("font", family="Malgun Gothic")
plt.title("키와 몸무게의 상관관계")
plt.scatter(height, weight, c=height, cmap="RdPu") # c: 색상의 진함,연함 강도
plt.xlabel("키")
plt.ylabel("몸무게")
plt.show()
```

```python
import random

height = []
weight = []

for i in range(100):
    height.append(random.randint(100, 200))
    weight.append(random.randint(20, 100))

plt.rc("font", family="Malgun Gothic")
plt.title("키와 몸무게의 상관관계")
plt.scatter(height, weight, c=height, cmap="RdPu") # c: 색상의 진함,연함 강도
plt.colorbar(label="키")    # 그래프 옆에 숫자별 색상값을 타나냄
plt.xlabel("키")
plt.ylabel("몸무게")
plt.show()
```

### 점의 크기 조절하기

```python
import random

height = []
weight = []
score = []

for i in range(100):
    height.append(random.randint(100, 200))
    weight.append(random.randint(20, 100))
    score.append(random.randint(1, 1000))

plt.rc("font", family="Malgun Gothic")
plt.title("키와 몸무게의 상관관계")
plt.scatter(height, weight, s=score, c=score, cmap="rainbow", alpha=0.7)  # c: 색상의 진함,연함 강도
plt.xlabel("키")
plt.ylabel("몸무게")
plt.colorbar(label="체육점수")    # 그래프 옆에 숫자별 색상값을 타나냄
plt.show()
```
